```
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```
genre_list= ['Rock', 'Pop', 'Rap', "Instrumental", "R&B", "Alternative", "Folk", "Jazz", "Reggae", "Soul"]
for genre in genre_list:
  df = pytrend.interest_by_region(resolution= 'COUNTRY', inc_low_vol=True, inc_geo_code=False)
  pytrend.build_payload(kw_list=[genre]) 
  df.head(10)
```
